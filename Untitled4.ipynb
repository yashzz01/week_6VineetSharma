```python
# MNIST Image Reconstruction Using Autoencoders

## Objective

The objective of this experiment is to implement and compare different Autoencoder architectures for image reconstruction using the MNIST handwritten digit dataset. The models learn compressed representations of images and reconstruct them while preserving important visual features.

## Introduction

Autoencoders are unsupervised neural networks that learn efficient representations of input data. They consist of two main components:

1. Encoder – Compresses the input into a lower-dimensional representation.
2. Decoder – Reconstructs the original input from the compressed representation.

The primary goal of an Autoencoder is to minimize the difference between the input image and the reconstructed image.

Autoencoders are widely used in:

* Image Compression
* Image Denoising
* Feature Extraction
* Dimensionality Reduction
* Anomaly Detection

## Dataset Description

The MNIST dataset consists of handwritten digit images ranging from 0 to 9.

Dataset Characteristics:

* Training Images: 60,000
* Testing Images: 10,000
* Image Size: 28 × 28 pixels
* Pixel Range: 0 to 255
* Grayscale Images

For model training, the images are normalized to values between 0 and 1.

## Methodology

### Data Preprocessing

1. Load the MNIST dataset.
2. Normalize pixel values between 0 and 1.
3. Split the training data into training and validation datasets.
4. Feed images to the Autoencoder model.

The target output is identical to the input image because Autoencoders learn reconstruction.

## Model 1: Feed Forward Neural Network (FFNN) Autoencoder

### Architecture

Encoder:

* Input Layer: 784 neurons
* Hidden Layer: Compressed representation

Decoder:

* Hidden Layer
* Output Layer: 784 neurons

### Working

The input image is flattened into a vector of length 784. The encoder compresses the image into a smaller representation, and the decoder reconstructs the image from this compressed representation.

### Advantages

* Simple implementation
* Fast training
* Easy to understand

### Limitations

* Ignores spatial information in images
* Lower reconstruction quality

## Model 2: Transpose CNN Autoencoder

### Architecture

Encoder:

* Convolution Layers
* Max Pooling Layers

Decoder:

* Transposed Convolution Layers
* Reconstruction Layer

### Working

The encoder extracts image features using convolution operations. The decoder reconstructs the image using transposed convolution layers, which increase image dimensions.

### Advantages

* Better feature extraction
* Improved reconstruction quality

### Limitations

* May produce checkerboard artifacts
* Higher computational complexity

## Model 3: Upsampled CNN Autoencoder

### Architecture

Encoder:

* Convolution Layers
* Pooling Layers

Decoder:

* Nearest Neighbor Upsampling
* Convolution Layers

### Working

Instead of transposed convolutions, the decoder first enlarges feature maps using upsampling and then applies convolution operations to reconstruct the image.

### Advantages

* Reduced checkerboard artifacts
* Stable image reconstruction
* Better visual quality

### Limitations

* Slightly higher computational cost

## Loss Function

### Mean Squared Error (MSE)

The reconstruction task is a regression problem because the network predicts pixel intensity values.

The Mean Squared Error loss is calculated as:

MSE = (1/N) Σ (Actual Pixel − Predicted Pixel)²

MSE measures the difference between original and reconstructed pixel values.

Reason for using MSE:

* Suitable for continuous numerical values.
* Measures reconstruction accuracy.
* Penalizes larger errors more heavily.
* Commonly used in image reconstruction tasks.

## Results

### Training Performance

[Insert Training Loss Graph Screenshot]

### Reconstruction Examples

[Insert Original vs Reconstructed Image Screenshot]

The reconstructed images successfully preserve the major visual characteristics of the original handwritten digits.

## Observations

1. FFNN Autoencoder successfully reconstructed images but produced relatively blurry outputs.
2. CNN-based Autoencoders preserved spatial information more effectively.
3. Transposed CNN Autoencoder achieved better reconstruction quality than FFNN.
4. Upsampled CNN Autoencoder reduced reconstruction artifacts and generated smoother images.
5. CNN Autoencoders learned meaningful feature representations of handwritten digits.
6. MSE loss decreased steadily during training, indicating successful learning.

## Comparison of Models

| Feature                  | FFNN Autoencoder | Transpose CNN Autoencoder | Upsampled CNN Autoencoder |
| ------------------------ | ---------------- | ------------------------- | ------------------------- |
| Complexity               | Low              | Medium                    | Medium                    |
| Training Speed           | Fast             | Moderate                  | Moderate                  |
| Spatial Feature Learning | Poor             | Good                      | Excellent                 |
| Reconstruction Quality   | Basic            | Good                      | Best                      |
| Artifact Reduction       | Low              | Medium                    | High                      |

## Conclusion

In this experiment, three Autoencoder architectures were implemented for MNIST image reconstruction. The FFNN Autoencoder demonstrated the basic concept of dimensionality reduction and reconstruction but lacked the ability to capture spatial image features. The Transpose CNN Autoencoder improved reconstruction quality through convolutional feature extraction but was susceptible to reconstruction artifacts. The Upsampled CNN Autoencoder produced the best visual results by combining upsampling and convolution operations, resulting in smoother and more accurate image reconstruction. Overall, CNN-based Autoencoders outperformed the FFNN Autoencoder in preserving image details and reconstruction quality.

## Learning Outcomes

* Understood the architecture and working of Autoencoders.
* Learned image compression and reconstruction techniques.
* Implemented FFNN and CNN-based Autoencoders.
* Explored latent space representations.
* Understood the role of MSE loss in reconstruction tasks.
* Compared different Autoencoder architectures using MNIST images.
```
